```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("/content/ABC Company.xlsx")
df
```

```python
import numpy as np
df['Height']=np.random.randint(150,180,size=len(df))
df
```

```python
dist=df['Team'].value_counts().reset_index()
dist.columns=['Team','Number of Employees']
dist['percentage']=(dist['Number of Employees']/len(df)*100)
print(dist)
plt.bar(x=dist['Team'],height=dist['Number of Employees'])
plt.title("Employee Distribution over Teams")
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.xticks(rotation=90)
plt.show()
plt.figure(figsize=(10,10))
plt.pie(dist['percentage'],labels=dist['Team'],autopct="%%1.1%%f")
plt.title("Percentage")
plt.show()
```

```python
#employee distibution is balanced with most of the companies have employees between 14 to 19.
```

```python
post=df['Position'].value_counts().reset_index()
post.columns=['Position','Number of Employees']
print(post)
plt.bar(x=post['Position'],height=post['Number of Employees'])
plt.title("Employee based on position")
plt.xlabel("Position")
plt.ylabel("Number of Employees")
plt.show()
```

```python
#the bar chart shows the highest number of people are employed in the position of SG and least number of people employed sa C in the above companies.
```

```python
age_counts=df['Age'].value_counts().reset_index()
age_counts.columns=['Age','Number of Employees']
predominant=age_counts.sort_values(by='Number of Employees',ascending=False)
display(predominant.head())
sns.barplot(x="Age",y="Number of Employees",data=age_counts.sort_values(by="Number of Employees",ascending=False),palette="viridis")
plt.xlabel("Age")
plt.ylabel("Number of Employees")
plt.title("Distribution of employees by Age")
plt.show()
```

```python
#This graph shows that there are 47 employees in the age group of 24.the age groups are between 23 and 27.
```

```python
expenditure=df.groupby(['Team','Position'])['Salary'].sum().reset_index()
highest=expenditure.sort_values(by='Salary',ascending =False)
print("Team and Position with highest salary")
display(highest.head())
sns.barplot(x='Team',y='Salary',data=highest.head(),palette="viridis")
plt.title("Team with highest salary")
plt.show()
sns.barplot(x='Position',y='Salary',data=highest.head(),palette="viridis")
plt.title("Position with highest salary")
plt.show()
```

```python
#the salary expenditure is highest for Los Angeles Lakers in the position of SF.
```

```python
correlation=df["Age"].corr(df["Salary"])
print("Correlation between age and salary is:",correlation)
sns.scatterplot(x="Age",y="Salary",data=df)
plt.title("Age vs Salary")
plt.show()
```

```python
#from the analysis of correlation between age and salary is a weak correlation of 0.21
```
